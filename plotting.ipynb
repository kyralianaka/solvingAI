```python
from model import solvingAI
import numpy as np
import matplotlib.pyplot as plt
```

```python
ei = solvingAI()
soln, weights_traj, weights_0, w = ei.run_sim()

im = plt.imshow(w)

plt.xlabel('Presynaptic Neurons')
plt.ylabel('Postsynaptic Neurons')
plt.colorbar(im)
plt.show()
```

```python

h_e = soln[:ei.N - ei.D, :]
h_i = soln[ei.N - ei.D:, :]
e_cur = w[: ei.N - ei.D, :ei.N - ei.D] @ ei.activate(h_e)
i_cur = - w[: ei.N - ei.D, ei.N - ei.D:] @ ei.activate(h_i) + 3
s = 400
for i in range(100):
    plt.plot(e_cur[i+10, :][-s:], c = 'r', alpha = 0.02)
    plt.plot(i_cur[-i-10, :][-s:], c = 'b', alpha = 0.02)

plt.plot(e_cur.mean(axis=0)[-s:], c = 'r', linewidth = 2)
plt.plot(i_cur.mean(axis=0)[-s:], c = 'b', linewidth = 2)
plt.show()
```
